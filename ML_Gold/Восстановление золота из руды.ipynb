# Восстановление золота из руды

Необходимо подготовить прототип модели машинного обучения для компании «Цифра». Компания разрабатывает решения для эффективной работы промышленных предприятий.

Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Предоставлены данные с параметрами добычи и очистки.

Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

**Этапы производства:**  
Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.



**1. Флотация**

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов. На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

**2. Очистка**

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

**Исходя из схемы, скорее всего потребуется разделить обучающую выборку на две части - флотации и два этапа очистки**

### Проверка расчета эффективности обогащения

Проверим, правильно ли рассчитана эффективность обогащения в обучающей выборке, рассчитаем ее на train. Найдем MAE между нашими расчётами и значением признака.  

Эффективность обогащения рассчитывается по **формуле**:



где:

C — доля золота в концентрате после флотации/очистки;  
F — доля золота в сырье/концентрате до флотации/очистки;  
T — доля золота в отвальных хвостах после флотации/очистки.  
Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

```python
# напишем функцию для расчета recovery
def recovery(df):
    C = df['rougher.output.concentrate_au']
    F = df['rougher.input.feed_au']
    T = df['rougher.output.tail_au']
    
    recovery = (((C * (F-T))) / (F * (C-T)))*100
    return recovery
```

```python
# создадим отдельный датасет с требуемыми метриками
tr_recovery = train[['rougher.output.concentrate_au', 
                    'rougher.input.feed_au', 
                    'rougher.output.tail_au',
                    'rougher.output.recovery']].copy()
# удалим пропущенные значения
tr_recovery = tr_recovery.dropna(axis=0)
# применим нашу функцию
tr_recovery['recovery_calculated'] = tr_recovery.apply(recovery, axis=1)
tr_recovery.head()
```

```python
# посчитаем МАЕ

print('Средняя абсолютная ошибка: {:.17f}'
      .format(mean_absolute_error(tr_recovery['rougher.output.recovery'], tr_recovery['recovery_calculated']))
     )
```

MAE стремится к нулю, поэтому считаем, что эффективность обогощения посчитана верно. Полученное значение оказалось в достаточной степени маленьким, чтобы можно было утверждать о правильности рассчёта.

### Анализ распределения размеров гранул сырья

Анализ распределения проведем на обучающей и тестовой выборках. Если распределения сильно отличаются друг от друга, оценка модели будет неправильной.

```python
# посмотрим какие столбцы потребуются для анализа 
df.loc[:, df.columns.str.contains('size')].columns
```

Посмотрим как распределны значения размера гранул при флотации.

```python
plt.figure(figsize=(14,6))
sns.kdeplot(train['rougher.input.feed_size'].dropna(), label='train_feed_size')
sns.kdeplot(test['rougher.input.feed_size'].dropna(), label='test_feed_size')
plt.title('Распределения размеров гранул сырья в обучающей и тестовой выборках')
plt.legend()
plt.show()
```

Вообще не обязательно проводить стат.тесты, достаточно анализа графика, но все-таки проведем его. Необходимые выводы можно сделать по графику, в нашем случае он дает более полную картину.

```python
# укажем уровень значимости
alpha = .05
```

```python
df_1 = train['rougher.input.feed_size'].dropna()
df_2 = test['rougher.input.feed_size'].dropna()

_, pval = st.levene(df_1, df_2)
print('p-value=%.3f' % pval)

if pval < alpha:
    print('Отвергаем Н0, дисперсии различаются')
else:
    print("Не отвергаем Н0, дисперсии не различаются")

var_df = [np.var(x, ddof=1) for x in [df_1, df_2]]
print('Дисперсии по выборкам: ', var_df)
```

Такие различия не помешают обучению, если посмотреть на график, то мы увидим - то что есть в тесте есть и в трейне и модель более менее равномерно обучится на всех диапазонах значений, которые есть в тесте. Получается у модели не должно быть сложностей при предсказании, по причине того, что в тесте встретится неизвестное ранее значение.

### Функция для вычисления итоговой sMAPE

Для решения задачи введём новую метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).  
Она похожа на MAE, но выражается не в абсолютных величинах, а в относительных. Она одинаково учитывает масштаб и целевого признака, и предсказания.  
Метрика sMAPE вычисляется так:  


Итоговая метрика складывается из двух величин:



```python
def smape(y_true, y_predict):
    y_true = np.array(y_true)
    y_predict = np.array(y_predict)
    total = 0
    for i in range(len(y_true)):
        total += abs(y_true[i] - y_predict[i]) / ((abs(y_true[i]) + abs(y_predict[i]))/2)
        smape = abs(total*100 / len(y_true))
    return smape
```

```python
score = make_scorer(smape, greater_is_better=False)
```

```python
def smape_total(smape_rougher, smape_final):
    smape_total = (0.25 * smape_rougher + 0.75 * smape_final)
    return smape_total
```
